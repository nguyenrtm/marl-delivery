```python
from env import Environment
from final import AStarAwareAgents as Agents
```

```python
map = "/kaggle/input/mapd01/marl-deliveryzip/map1.txt"
max_time_steps = 1000
n_robots = 5
n_packages = 100
seed = 2025
```

```python
import numpy as np

num=20
results = []
for i in range(num):
    env = Environment(map_file=map, max_time_steps=max_time_steps,
                      n_robots=n_robots, n_packages=n_packages,
                      seed = seed)
    state = env.reset()
    agents = Agents()
    agents.init_agents(state)
    done = False
    t = 0
    while not done:
        actions = agents.get_actions(state)
        next_state, reward, done, infos = env.step(actions)
        state = next_state
        t += 1
    results.append(infos["total_reward"])
avg_reward = np.mean(results)
std_reward = np.std(results)
print(f"Results: {results}")
print(f"Average Reward: {avg_reward}")
print(f"Standard Deviation of Reward: {std_reward}")
```
